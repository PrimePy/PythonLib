```python
import re
import pandas as pd

# Marval movie data lists
movie_name=[]
movie_release_year=[]
movie_rating=[]

moviesDescription=html.find_all("div",class_="list-description")
i=0
for movieDescription in moviesDescription:
    if movieDescription.find("b") is not None:
         # list of movie name
        movie_name.append(movieDescription.a.text.strip())
        
        movie_details=movieDescription.b.text.strip()
        movie_data=movie_details.split(" ")
        
        # list of movie release year
        movie_release_year.append(int(re.sub("[\(\)]",'',movie_data[-4])))
        
        # list of movie rating
        movie_rating.append(float(movie_data[-1]))

dataframe=pd.DataFrame({
    "Name":movie_name,
    "ReleaseYear":movie_release_year,
    "Rating":movie_rating
})

print(dataframe)
```
