```python
import pandas as pd
```

```python
a=pd.DataFrame({
    'city':['chennai','trichy','pudukkottai','tanjavur','namakkal'],
    'tempearture':[23,54,65,32,98]
})
a
```

```python
b=pd.DataFrame({
    'city':['chennai','trichy','pudukkottai','madurai'],
    'humidity':[23,12,54,87]
})
b
```

```python
pd.merge(a,b,on='city')
```

```python
pd.merge(a,b,on='city',how='inner')
```

```python
pd.merge(a,b,on='city',how='outer')
```

```python
pd.merge(a,b,on='city',how='left')
```

```python
pd.merge(a,b,on='city',how='right')
```

```python
pd.merge(a,b,on='city',how='outer',indicator=True)
```
