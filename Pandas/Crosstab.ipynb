```python
import pandas as pd
df=pd.read_excel('survey.xlsx')
df
```

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Nationality,df.Handedness,margins=True)
```

```python
 pd.crosstab(df.Nationality,[df.Handedness,df.Sex],margins=True)
```

```python
pd.crosstab([df.Sex,df.Handedness],df.Nationality,margins=True)
```

```python
pd.crosstab(df.Nationality,df.Sex,margins=True)
```

```python
pd.crosstab(df.Nationality,df.Sex,normalize='index', margins=True)
```

```python
import numpy as np
pd.crosstab(df.Nationality,df.Handedness,values=df.Age,aggfunc=np.average)
```
